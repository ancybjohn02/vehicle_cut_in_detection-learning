```python
import cv2
imag1 = cv2.imread("tribe people fancy dress.JPG",1)
# imag1 = cv2.imread("tribe people fancy dress.JPG",0)
#  the second parameter is for color, by default color, 0 for grayscale and 1 for color
imag1 = cv2.resize(imag1,(1280,700))
print(imag1)
cv2.imshow("original", imag1)

k = cv2.waitKey()
if k ==ord("s") :
    cv2.imwrite("/home/zeal/programming/computer_vision/tutorial/img1_output.jpeg")
else :
    # waitkey(5000) the img window will diappear after 5 sec
    cv2.destroyAllWindows()

imag2 = cv2.imread("tribe people fancy dress.JPG",0)
imag2 = cv2.resize(imag2,(1280,700))
cv2.imshow("grayscale img", imag2)
if k ==ord("s") :
    cv2.imwrite("/home/zeal/programming/computer_vision/tutorial/img2_output.jpeg")
else :
    # waitkey(5000) the img window will diappear after 5 sec
    cv2.destroyAllWindows()

imag3 = cv2.imread("tribe people fancy dress.JPG",-1)
imag3 = cv2.resize(imag3,(1280,700))
cv2.imshow("unchanged", imag3)
# here saturation will be a little more, alpha channel will b a lil more

if k ==ord("s") :
    cv2.imwrite("/home/zeal/programming/computer_vision/tutorial/img3_output.jpeg")
else :
    # waitkey(5000) the img window will diappear after 5 sec
    cv2.destroyAllWindows()



# flip image
img4 = cv2.flip(imag1,0) #can accept 3 param : 0,-1,1
cv2.imshow("converted image :", img4)
```
