#### same code, just in another way

```python
import cv2

def display_and_save(image, title):
  """Displays an image with a title and prompts for saving."""
  cv2.imshow(title, image)
  save_query = input("Save all images (y/n)? ").lower()
  if save_query == 'y':
    filename_base = input("Enter base filename (default: output): ") or "output"
    cv2.imwrite(f"{filename_base}_original.jpg", image.copy())
    cv2.imwrite(f"{filename_base}_grayscale.jpg", cv2.cvtColor(image.copy(), cv2.COLOR_BGR2GRAY))
    cv2.imwrite(f"{filename_base}_flipped_horizontal.jpg", cv2.flip(image.copy(), 0))
    cv2.imwrite(f"{filename_base}_flipped_vertical.jpg", cv2.flip(image.copy(), 1))
  cv2.waitKey(0)  # Wait indefinitely for a key press
  cv2.destroyAllWindows()

def main():
  # Read the image
  image = cv2.imread("tribe people fancy dress.JPG")

  # Get user input for resizing
  width = int(input("Enter desired width (default: 1280): ") or 1280)
  height = int(input("Enter desired height (default: 700): ") or 700)

  # Resize the image based on user input
  resized_image = cv2.resize(image, (width, height))

  # Display options for user
  print("Choose an option:")
  print("1. Show Original Image")
  print("2. Convert to Grayscale")
  print("3. Flip Horizontally")
  print("4. Flip Vertically")
  print("5. Show All (Original, Grayscale, Flipped)")
  choice = int(input("Enter your choice (1-5): "))

  if choice == 1:
    display_and_save(resized_image.copy(), "Original Image")
  elif choice == 2:
    grayscale_image = cv2.cvtColor(resized_image.copy(), cv2.COLOR_BGR2GRAY)
    display_and_save(grayscale_image, "Grayscale Image")
  elif choice == 3:
    flipped_horizontally = cv2.flip(resized_image.copy(), 0)
    display_and_save(flipped_horizontally, "Flipped Horizontally")
  elif choice == 4:
    flipped_vertically = cv2.flip(resized_image.copy(), 1)
    display_and_save(flipped_vertically, "Flipped Vertically")
  elif choice == 5:
    # Show all images in separate windows
    cv2.imshow("Original Image", resized_image.copy())
    cv2.imshow("Grayscale Image", cv2.cvtColor(resized_image.copy(), cv2.COLOR_BGR2GRAY))
    cv2.imshow("Flipped Horizontally", cv2.flip(resized_image.copy(), 0))
    cv2.imshow("Flipped Vertically", cv2.flip(resized_image.copy(), 1))
    display_and_save(resized_image.copy(), "All Images (Press any key to close)")  # Dummy call for saving prompt
  else:
    print("Invalid choice!")

if __name__ == "__main__":
  main()
```

```python
# input from webcam 
import cv2

cap = cv2.VideoCapture(0) # pass 1 for external camera
print("Capture", cap)

fourcc = cv2.VideoWriter_fourcc("XVID")
output = cv2.VideoWriter("output.avi", fourcc,20.0,(640,480),0) # the last 0 specifies, gray or color
# param = name,codec,fps, resolution 

while cap.isOpened():
    ret, frame = cap.read()
    if ret == True:
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frame = cv2.flip(frame,0)
        cv2.imshow("frame", frame)
        cv2.imshow("gray",gray)
        k = cv2. waitKey(25) #this no. inside waitkey controls frames/time, so playback speed of vdo
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
cap.release()
output.release() #otherwise vdo can be corrupted
cv2.destroyAllWindows()
```
